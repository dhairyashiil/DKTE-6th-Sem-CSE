```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a=sns.load_dataset("flights")
sns.relplot(x="passengers",y="month",hue="year",data=a)
```

```python
b=sns.load_dataset("tips")
sns.relplot(x="time",y="tip",data=b,kind="line" )
```

```python
sns.relplot(x="total_bill",y="tip",data=b,hue="smoker")
```

```python
sns.relplot(x="total_bill",y="tip",data=b,hue="smoker",style="smoker")
```

```python
sns.relplot(x="total_bill",y="tip",data=b,hue="smoker",style="time")
```

```python
sns.relplot(x="total_bill",y="tip",data=b,hue="size")
```

# For Categorical Variable

```python
sns.set(style="ticks",color_codes=True)
tips=sns.load_dataset("tips")
```

```python
sns.catplot(x="day",y="total_bill",data=tips)
```

```python
sns.catplot(x="day",y="total_bill",data=tips,jitter=False)
```

```python
sns.catplot(x="day",y="total_bill",data=tips,kind="swarm")
```

```python
sns.catplot(x="day",y="total_bill",data=tips,kind="strip")
```

```python
sns.catplot(x="day",y="total_bill",data=tips,kind="swarm",hue="sex")
```
