```python
df3 = pd.read_excel('Anova_Dataset.xlsx')
df3
```

```python
sns.boxplot(x='Power', y='GrowthRate', data=df3)
```

Change the Persepective of data

```python
df3_pivot = df3.pivot(index='RunID', columns='Power', values='GrowthRate')
df3_pivot
```

```python
df3_pivot.sum()
```

```python
def degrees_of_freedom(df3_pivot):
    ## Between samples is across the row
    ## Within samples is down the column
    n = df3_pivot.count(axis=1) # Define number of replicates within treatment
    a = df3_pivot.count() # Determine number of treatments
    n = n.iloc[0]
    a = a.iloc[0]
    
    TotalN = n*a # Total number of samples
    dft = a - 1 #dft is degrees of freedom between sample set
    dfe = TotalN - a #dfe is degrees of freedom within samples
    dftotal = TotalN-1 #dftotal is degrees of freedom total 
    return dft, dfe, dftotal,n, a
```

```python
dft, dfe, dftotal,n,a = degrees_of_freedom(df3_pivot)

print("Degrees of freedom between treatments: {:d}".format(dft))
print("Degrees of freedom within treatments: {:d}".format(dfe))
print("Total degrees of freedom: {:d}".format(dftotal))
```

```python
def compute_sums(df3_pivot):
    yi_sum = df3_pivot.sum() # Compute sum under i-th treatment
    yi_avg = df3_pivot.mean() # Compute average under i-th treatment 
    y_sum = yi_sum.sum() # Compute grand sum
    y_avg = yi_avg.mean() # Compute grand average
    return yi_sum, yi_avg, y_sum, y_avg

yi_sum, yi_avg, y_sum, y_avg = compute_sums(df3_pivot) #retrieve respective sums by calling function

print(r"The sum under each Power (W) treatment is: {}".format(yi_sum))
print(r"The average under each Power (W) treatment is: {}".format(yi_avg))
print(r"The grand sum is: {:.2f}".format(y_sum))
print(r"The grand average is: {:.2f}".format(y_avg)) 
```

```python
def Sum_of_Squares(yi_avg, y_avg, df3_pivot):
    SSTR = (yi_avg - y_avg)**2 #Calculate sum of squares between treatments
    SSTR = SSTR.sum()*n
    SSE = df3_pivot.sub(yi_avg.values) ** 2 #Calculate sum of squares within treatments
    SSE = SSE.sum().sum()
    SST = SSTR + SSE # Total corrected sum of squares
    return SSTR, SSE, SST

SSTR, SSE, SST = Sum_of_Squares(yi_avg,y_avg, df3_pivot)

print("Sum of Squares between treatments is: {:.2f}".format(SSTR))
print("Sum of Squares within treatments is: {:.2f}".format(SSE))
print("Total Sum of Squares is: {:.2f}".format(SST))
```

Correlation: a measure of the extent of interdependence between variables.

Causation: the relationship between cause and effect between two variables.

Pearson Correlation

    The Pearson Correlation measures the linear dependence between two variables X and Y.
    
    The resulting coefficient is a value between -1 and 1 inclusive, where:

    1: Total positive linear correlation.
    0: No linear correlation, the two variables most likely do not affect each other.
    -1: Total negative linear correlation.

# 5.ANOVA (F-Test)

The Analysis of Variance (ANOVA) is a statistical method used to test whether there are significant differences between the means of two or more groups. ANOVA returns two parameters:

F-test score: ANOVA assumes the means of all groups are the same, calculates how much the actual means deviate from the assumption, and reports it as the F-test score. A larger score means there is a larger difference between the means.

P-value: P-value tells how statistically significant is our calculated score value.

If our price variable is strongly correlated with the variable we are analyzing, expect ANOVA to return a sizeable F-test score and a small p-value.

```python
df_gptest = data[['drive-wheels','body-style','price']]
grouped_test2=df_gptest[['drive-wheels', 'price']].groupby(['drive-wheels'])
grouped_test2.head()
```

```python
grouped_test2.get_group('4wd')['price']
```

```python
#Anova
f_value,p_value= stats.f_oneway(grouped_test2.get_group('fwd')['price'],grouped_test2.get_group('rwd')['price'],grouped_test2.get_group('4wd')['price'])
print( "ANOVA results: F=", f_value, ", P =", p_value) 

#Generally used for the categorical data
```

This is a great result, with a large F test score showing a strong correlation
